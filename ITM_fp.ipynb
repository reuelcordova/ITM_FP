```python
new_df = df.loc[:,['Total Household Income','Region', 'Total Food Expenditure','Bread and Cereals Expenditure', 'Total Rice Expenditure',
       'Meat Expenditure', 'Total Fish and  marine products Expenditure',
       'Fruit Expenditure', 'Vegetables Expenditure',
       'Restaurant and hotels Expenditure', 'Alcoholic Beverages Expenditure',
       'Tobacco Expenditure', 'Clothing, Footwear and Other Wear Expenditure',
       'Housing and water Expenditure', 'Imputed House Rental Value',
       'Medical Care Expenditure', 'Transportation Expenditure',
       'Communication Expenditure', 'Education Expenditure',
       'Miscellaneous Goods and Services Expenditure','Special Occasions Expenditure', 'Crop Farming and Gardening expenses','Household Head Sex','Household Head Highest Grade Completed',
       'Household Head Occupation','Household Head Class of Worker', 
]]
```

# Sorting and Filtering the Data

```python
new_df['Vice Expenditures'] = df.loc[:,['Alcoholic Beverages Expenditure','Tobacco Expenditure']].sum(axis=1)
new_df['Discretionary Expenses'] = df.loc[:,['Restaurant and hotels Expenditure','Special Occasions Expenditure']].sum(axis=1)
new_df['New Total Food Expenditure'] = df.loc[:,['Bread and Cereals Expenditure', 'Total Rice Expenditure',
       'Meat Expenditure', 'Total Fish and  marine products Expenditure',
       'Fruit Expenditure', 'Vegetables Expenditure',]].sum(axis=1)
new_df = new_df.drop(['Bread and Cereals Expenditure', 'Total Rice Expenditure',
       'Meat Expenditure', 'Total Fish and  marine products Expenditure',
       'Fruit Expenditure', 'Vegetables Expenditure','Alcoholic Beverages Expenditure',
       'Tobacco Expenditure','Restaurant and hotels Expenditure','Special Occasions Expenditure'], axis = 1)
new_df = new_df.sort_values('Total Household Income')
```

```python
rating = []
for row in new_df['Household Head Highest Grade Completed']:
    if row == 'Preschool':
        rating.append('Preschool')
    elif row == 'Grade 1':
        rating.append('Preschool')
    elif row == 'Grade 2':
        rating.append('Preschool')
    elif row == 'Grade 3':
        rating.append('Preschool')
    elif row == 'Grade 4':
        rating.append('Preschool')
    elif row == 'Grade 5':
        rating.append('Preschool')
    elif row == 'Grade 6':
        rating.append('Preschool')
    elif row == 'First Year High school':
        rating.append('Grade School')
    elif row == 'Second Year High School':
        rating.append('Grade School')
    elif row == 'Third Year High School':
        rating.append('Grade School')
    elif row == 'First Year College':
        rating.append('High School')
    elif row == 'Second Year College':
        rating.append('High School')
    elif row == 'Third Year College':
        rating.append('High School')
    elif row == 'Elementary Graduate':
        rating.append('Grade School')
    elif row == 'High School Graduate':
        rating.append('High School')
    elif row == 'No Grade Completed':
        rating.append('No Grade Completed')
    else:
        rating.append('College and up')
        
new_df['New Highest Household Head Grade Completed'] = rating
```

### Number of Automobiles in NCR, Calabarzon, Central Luzon, Davao Region

```python
def car_sum(place):
    return df.loc[df['Region'] == place]['Number of Car, Jeep, Van'].sum()

def motor_sum(place):
    return df.loc[df['Region'] == place]['Number of Motorcycle/Tricycle'].sum()

def boat_sum(place):
    return df.loc[df['Region'] == place]['Number of Motorized Banca'].sum()

cars = [car_sum('NCR'),car_sum('IVA - CALABARZON'),car_sum('III - Central Luzon'),car_sum('XI - Davao Region')]
motor = [motor_sum('NCR'),motor_sum('IVA - CALABARZON'),motor_sum('III - Central Luzon'),motor_sum('XI - Davao Region')]
boat = [boat_sum('NCR'),boat_sum('IVA - CALABARZON'),boat_sum('III - Central Luzon'),boat_sum('XI - Davao Region')]
```

```python
auto_plot = pd.DataFrame({'Number of Cars/Jeep/Van': cars,'Number of Motorcycles/Tricycles': motor,'Number of motorized boats': boat},index = ['NCR','Calabarzon','Central Luzon','Davao Region'])
auto_plot.plot(kind = 'bar',stacked = True, color = ['#3cafb7','#aab73c','#2bc372'])
plt.xlabel ('Region')
plt.ylabel ('Number of Cars')
plt.title ('Number of Automobiles in NCR, Calabarzon, Central Luzon and Davao Region')
plt.show ()
```

## Expenditures of Male v Female

```python
exp_df = new_df.loc[:,['New Total Food Expenditure',
        'Clothing, Footwear and Other Wear Expenditure',
       'Housing and water Expenditure','Medical Care Expenditure', 'Transportation Expenditure',
       'Communication Expenditure', 'Education Expenditure',
       'Miscellaneous Goods and Services Expenditure', 'Crop Farming and Gardening expenses',
        'Discretionary Expenses','Vice Expenditures','Household Head Sex']]
```

```python
m_exp = exp_df.loc[(exp_df['Household Head Sex'] == 'Male')]
f_exp = exp_df.loc[(exp_df['Household Head Sex'] == 'Female')]
```

```python
def exp_average(df):
    average = []
    for col in df:
        if col != 'Household Head Sex':
            average.append(df[col].mean())
    return average
```

```python
expenditure = []
for col in exp_df:
    if col != 'Household Head Sex':
        expenditure.append(col)
```

```python
df1=pd.DataFrame({'x':expenditure,'y':exp_average(m_exp)})
df2=pd.DataFrame({'x':expenditure,'y':exp_average(f_exp)})
df1['hue']= 'Male'
df2['hue']= 'Female'
res=pd.concat([df1,df2])
sns.barplot(x='y',y='x',data=res,hue='hue',palette = 'YlGnBu')
plt.show()
```
