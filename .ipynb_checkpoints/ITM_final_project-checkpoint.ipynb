```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv('Family-Income-and-Expenditure.csv')
```

# Sorting and Filtering the Data

```python
new_df = df.loc[:,['Total Household Income','Region', 'Total Food Expenditure','Bread and Cereals Expenditure', 'Total Rice Expenditure',
       'Meat Expenditure', 'Total Fish and  marine products Expenditure',
       'Fruit Expenditure', 'Vegetables Expenditure',
       'Restaurant and hotels Expenditure', 'Alcoholic Beverages Expenditure',
       'Tobacco Expenditure', 'Clothing, Footwear and Other Wear Expenditure',
       'Housing and water Expenditure', 'Imputed House Rental Value',
       'Medical Care Expenditure', 'Transportation Expenditure',
       'Communication Expenditure', 'Education Expenditure',
       'Miscellaneous Goods and Services Expenditure','Special Occasions Expenditure', 'Crop Farming and Gardening expenses','Household Head Sex','Household Head Highest Grade Completed',
       'Household Head Occupation','Household Head Class of Worker',
        'Number of Car, Jeep, Van','Number of Motorcycle/Tricycle', 'Number of Motorized Banca',
]]
new_df['Vice Expenditures'] = df.loc[:,['Alcoholic Beverages Expenditure','Tobacco Expenditure']].sum(axis=1)
new_df['Discretionary Expenses'] = df.loc[:,['Restaurant and hotels Expenditure','Special Occasions Expenditure']].sum(axis=1)
new_df['New Total Food Expenditure'] = df.loc[:,['Bread and Cereals Expenditure', 'Total Rice Expenditure',
       'Meat Expenditure', 'Total Fish and  marine products Expenditure',
       'Fruit Expenditure', 'Vegetables Expenditure',]].sum(axis=1)
new_df = new_df.drop(['Bread and Cereals Expenditure', 'Total Rice Expenditure',
       'Meat Expenditure', 'Total Fish and  marine products Expenditure',
       'Fruit Expenditure', 'Vegetables Expenditure','Alcoholic Beverages Expenditure',
       'Tobacco Expenditure','Restaurant and hotels Expenditure','Special Occasions Expenditure'], axis = 1)
new_df = new_df.sort_values('Total Household Income')
rating = []
for row in new_df['Household Head Highest Grade Completed']:
    if row == 'Preschool':
        rating.append('Preschool')
    elif row == 'Grade 1':
        rating.append('Preschool')
    elif row == 'Grade 2':
        rating.append('Preschool')
    elif row == 'Grade 3':
        rating.append('Preschool')
    elif row == 'Grade 4':
        rating.append('Preschool')
    elif row == 'Grade 5':
        rating.append('Preschool')
    elif row == 'Grade 6':
        rating.append('Preschool')
    elif row == 'First Year High school':
        rating.append('Grade School')
    elif row == 'Second Year High School':
        rating.append('Grade School')
    elif row == 'Third Year High School':
        rating.append('Grade School')
    elif row == 'First Year College':
        rating.append('High School')
    elif row == 'Second Year College':
        rating.append('High School')
    elif row == 'Third Year College':
        rating.append('High School')
    elif row == 'Elementary Graduate':
        rating.append('Grade School')
    elif row == 'High School Graduate':
        rating.append('High School')
    elif row == 'No Grade Completed':
        rating.append('No Grade Completed')
    else:
        rating.append('College and up')       
new_df['New Highest Household Head Grade Completed'] = rating
```

#### New columns:
- Vice Expenditures: Sum of Alcoholic Beverages Expenditure & Tobacco Expenditure
- Discretionary Expenses: Sum of restaurant & hotels expenditure and Special Occasions Expenditure
- New Total Food Expenditure: Sum of Bread & Cereal Expenditure, Total Rice Expenditure, Meat Expenditure, Total Fish and marine products expenditure, Fruit expenditure and Vegetable expenditure
- New Highest Household Head Grade Completed: Simplifies the data of Household head highest grade completed

#### Data is sorted from lowest total household income to highest household income

```python
auto_plot = pd.DataFrame({'Number of Cars/Jeep/Van': cars,'Number of Motorcycles/Tricycles': motor,'Number of motorized boats': boat},index = ['NCR','Calabarzon','Central Luzon','Davao Region'])
auto_plot.plot(kind = 'bar',stacked = True, color = ['#345590','#51a9b3','#deeaba'])
plt.xlabel ('Region')
plt.ylabel ('Number of Automobiles')
plt.title ('Number of Automobiles in NCR, Calabarzon, Central Luzon and Davao Region')
#plt.savefig('autostacked.png',bbox_inches='tight')
plt.show ()
```

### Expenditures of Male and Female Household

```python
exp_df = new_df.loc[:,['New Total Food Expenditure',
        'Clothing, Footwear and Other Wear Expenditure',
       'Housing and water Expenditure','Medical Care Expenditure', 'Transportation Expenditure',
       'Communication Expenditure', 'Education Expenditure',
       'Miscellaneous Goods and Services Expenditure', 'Crop Farming and Gardening expenses',
        'Discretionary Expenses','Vice Expenditures','Household Head Sex']]
m_exp = exp_df.loc[(exp_df['Household Head Sex'] == 'Male')]
f_exp = exp_df.loc[(exp_df['Household Head Sex'] == 'Female')]
def exp_average(df):
    average = []
    for col in df:
        if col != 'Household Head Sex':
            average.append(df[col].mean())
    return average
expenditure = []
for col in exp_df:
    if col != 'Household Head Sex':
        expenditure.append(col)
```

```python
df1=pd.DataFrame({'x':expenditure,'y':exp_average(m_exp)})
df2=pd.DataFrame({'x':expenditure,'y':exp_average(f_exp)})
df1['hue']= 'Male'
df2['hue']= 'Female'
res=pd.concat([df1,df2])
sns.barplot(x='y',y='x',data=res,hue='hue',palette = 'YlGnBu')
plt.title('Expenditures of Male and Female Household')
plt.xlabel('')
plt.ylabel("")
#plt.savefig('sex&expenditure',bbox_inches='tight')
plt.show()
```

## Expenditure and Social Class

```python
rating = []
for row in new_df['Total Household Income']:
    if row < 9100:
        rating.append("Poor")
    elif 9100 <= row < 18200:
        rating.append("Low Income")
    elif 18200 <= row < 36400:
        rating.append("Lower Middle")
    elif 36400 <= row < 63700:
        rating.append("Middle Middle")
    elif 63700 <= row < 109200:
        rating.append("Upper Middle")
    elif 109200 <= row < 182000:
        rating.append("Upper Class")
    elif 18200 <= row:
        rating.append("Rich")
new_df["Social Class"] = rating
```

```python
plt.rcParams['font.size'] = 10


for social_class in new_df["Social Class"].unique():
    df_class = new_df[new_df["Social Class"] == social_class]
    exp_df = df_class[['New Total Food Expenditure', 'Clothing, Footwear and Other Wear Expenditure', 'Housing and water Expenditure', 'Medical Care Expenditure', 'Transportation Expenditure', 'Communication Expenditure', 'Education Expenditure', 'Miscellaneous Goods and Services Expenditure', 'Crop Farming and Gardening expenses', 'Discretionary Expenses', 'Vice Expenditures']]
    total_expenditure = exp_df.sum()

    #create pie chart
    plt.figure(figsize=(8, 8))
    pie = plt.pie(total_expenditure, labels=None, colors=sns.color_palette('YlGnBu'), autopct=lambda p: '{:.0f}%'.format(p) if p >= 5 else '', labeldistance=1.1)

    #add legend
    plt.legend(pie[0], exp_df.columns, loc='center left', bbox_to_anchor=(1, 0.5))

    plt.title(f"{social_class} Household Expenditure")

plt.show()
```
